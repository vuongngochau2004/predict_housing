## Chuẩn hóa cột Giá bán thành đơn vị tỷ

Chuyển đổi giá bán từ format text (ví dụ: "1,5 tỷ", "500 triệu") sang số thực với đơn vị tỷ.

```python
def convert_price_to_billion(price_str):
    """
    Chuyển đổi giá từ text sang số với đơn vị tỷ
    Ví dụ: '1,5 tỷ' -> 1.5
           '500 triệu' -> 0.5
           '7 tỷ' -> 7.0
    """
    if pd.isna(price_str):
        return None
    
    # Xóa khoảng trắng thừa
    price_str = str(price_str).strip()
    
    try:
        if 'tỷ' in price_str:
            # Trường hợp tỷ: lấy số trước 'tỷ'
            number = price_str.replace('tỷ', '').strip()
            # Thay dấu phẩy thành dấu chấm
            number = number.replace(',', '.')
            return float(number)
        elif 'triệu' in price_str:
            # Trường hợp triệu: chia cho 1000 để chuyển sang tỷ
            number = price_str.replace('triệu', '').strip()
            number = number.replace(',', '.')
            return float(number) / 1000
        else:
            # Thử convert trực tiếp
            return float(price_str.replace(',', '.'))
    except:
        return None

# Áp dụng hàm chuyển đổi
df['Giá bán'] = df['Giá bán'].apply(convert_price_to_billion)
df.head()
```
